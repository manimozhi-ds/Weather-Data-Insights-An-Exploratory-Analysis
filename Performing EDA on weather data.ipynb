Analysis of relationship between variables

Correlation

```python
num_cols = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Visibility (km)']
correlation =data[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='Reds', square=True, annot_kws={'fontsize':10,'font':'Times New Roman'})
plt.xticks(font='Times New Roman', fontsize=10)
plt.yticks(font='Times New Roman', fontsize=10)
plt.title('Correlation Heatmap', font = 'Times New Roman')
plt.show()
```

Top 10 Weather Summary

```python
plt.figure(figsize=(14, 6))
top_summaries = data['Summary'].value_counts().nlargest(10)
sns.barplot(x=top_summaries.values, y=top_summaries.index, hue = top_summaries.index, legend = False, palette="coolwarm")
plt.title("Top 10 Weather Summary Conditions")
plt.xlabel("Count")
plt.ylabel("Summary")
plt.show()
```

Feature Engineering

```python
data['Formatted Date']=pd.to_datetime(data['Formatted Date'], utc=True)
```

```python
data['Month'] = data['Formatted Date'].dt.month
```

```python
data['Hour'] = data['Formatted Date'].dt.hour 
```

```python
hourly_temp = data.groupby('Hour')['Temperature (C)'].mean()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=hourly_temp.index,y=hourly_temp.values, marker='o', markersize=6, linestyle='--', linewidth=2, color='Purple')
plt.title("Average Temperature by Hour of Day", font='Times New Roman')
plt.xlabel("Hour",font='Times New Roman', fontsize=10)
plt.ylabel("Temperature (C)",font='Times New Roman', fontsize=10)
plt.xticks(ticks=range(0, 24), fontsize=6, font='Times New Roman')
plt.yticks(fontsize=6, font='Times New Roman')
plt.grid(True)
plt.show()
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(3, 2, figsize=(18, 16))
fig.suptitle('Advanced EDA of Weather Data', fontsize=20)

# 1. Temperature Distribution
sns.histplot(data['Temperature (C)'], bins=50, kde=True, ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Temperature Distribution')

# 2. Temperature vs Apparent Temperature
sns.scatterplot(x='Temperature (C)', y='Apparent Temperature (C)', data=data, ax=axs[0, 1], alpha=0.3)
axs[0, 1].set_title('Temperature vs Apparent Temperature')

# 3. Humidity vs Temperature
sns.scatterplot(x='Temperature (C)', y='Humidity', data=data, ax=axs[1, 0], alpha=0.3, color='purple')
axs[1, 0].set_title('Humidity vs Temperature')

# 4. Wind Speed vs Temperature
sns.scatterplot(x='Temperature (C)', y='Wind Speed (km/h)', data=data, ax=axs[1, 1], alpha=0.3, color='green')
axs[1, 1].set_title('Wind Speed vs Temperature')

# 5. Monthly Average Temperature

monthly_avg_temp = data.groupby('Month')['Temperature (C)'].mean()
sns.lineplot(x=monthly_avg_temp.index, y=monthly_avg_temp.values, ax=axs[2, 0], marker='o')
axs[2, 0].set_title('Average Monthly Temperature')
axs[2, 0].set_xlabel("Month")
axs[2, 0].set_ylabel("Temperature (C)")
axs[2, 0].set_xticks(range(1, 13))

# 6. Precipitation Type Distribution
sns.countplot(data=data, x='Precip Type',hue='Precip Type', legend=False, ax=axs[2, 1], palette='pastel')
axs[2, 1].set_title('Precipitation Type Distribution')


# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
